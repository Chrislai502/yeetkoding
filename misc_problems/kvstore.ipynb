```python
# Implement (code) a Key value store with transactions.

# Write a Fully funcitonal code in 25-30 min in interview with test cases

# Set
# Get
# Delete are methods in Key value store

# for transactions
# Begin
# Commit
# Rollback

class KVStore:
    def __init__(self):
        self.stack = [{}]
        
    def get(self, key):
        for i in range(len(self.stack) -1, -1, -1):
            # Get the first value seen
            if key in self.stack[i]:
                return self.stack[i][key]
    def set(self, key, value):
        self.stack[-1][key] = value
    def delete(self, key, values):
        for i in range(len(self.stack) -1, -1, -1):
            # Get the first value seen
            if key in self.stack[i]:
                del(self.stack[i][key])
    def begin(self):
        self.stack.append({})
    def commit(self):
        if len(self.stack) > 1:
            temp = self.stack.pop()
            for k, v in temp.items():
                self.stack[-1][k] = v
    def rollback(self):
        self.stack.pop()
        
# Test Transactions
# Add 3 values, with nested transactions
kv = KVStore()

kv.set(1, 11)
assert kv.get(1) == 11

kv.begin()
kv.set(1, 22)
kv.set(2, 22)
assert kv.get(1) == 22
assert kv.get(2) == 22

kv.begin()
kv.set(3, 33)
kv.set(2, 44)

assert kv.get(1) == 22
assert kv.get(2) == 44
assert kv.get(3) == 33

kv.commit()
kv.rollback()

assert kv.get(1) == 11
assert kv.get(2) == None
assert kv.get(3) == None
```
